```python
import csv
import pandas as pd

DE_folder = "$HOME/Bioinformatics_data/lferriphilum/data/RNA_data/DeSeq2/"
ann_folder = "$HOME/Bioinformatics_data/lferriphilum/analysis/DNA/04_genome_annotation/eggNOG-mapper_with_proteins/"

DE_file = "LFerr_condition_Batch_vs_Continuous_DeSeq2_results.csv"
ann_file = "LFerr_annotation.tsv"
```

```python
header = ["query_name", "seed_eggNOG_ortholog", "seed_ortholog_evalue", "seed_ortholog_score", "best_tax_level",
          "Preferred_name", "GOs", "EC", "KEGG_ko", "KEGG_Pathway", "KEGG_Module", "KEGG_Reaction", "KEGG_rclass",
          "BRITE", "KEGG_TC", "CAZy", "BiGG_Reaction", "Kingdom", "Clusters_of_Orthologous_Groups", "??", "???", "Description"]
ann = pd.read_csv(ann_folder + ann_file, sep='\t', skiprows=4, names=header)
```

```python
ann_subset = ann[['query_name', 'Preferred_name', 'Description']]
ann_subset
```

```python
header = ["prokka_gene_name", "baseMean", "log2FoldChange", "lfcSE", "stat", "pvalue", "padj"]
de = pd.read_csv(DE_folder + DE_file, sep=',', names=header, skiprows=1)
```

```python
de
```

```python
merged = pd.merge(ann_subset, de, how="right", left_on="query_name", right_on="prokka_gene_name")
```

```python
merged = merged[["prokka_gene_name", "Preferred_name", "Description", "baseMean", "log2FoldChange", "lfcSE", "stat", "pvalue", "padj"]]
merged = merged.sort_values("padj")
merged
```

```python
out_dir="$HOME/Bioinformatics_data/lferriphilum/data/RNA_data/DeSeq2/"
out_name="LFerr_condition_Batch_vs_Continuous_DeSeq2_results_with_annotation.csv"

merged.to_csv(out_dir + out_name, index=False)
```
